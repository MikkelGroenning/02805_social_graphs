# Part 1: Binning degree distributions

### Binning real numbers

**Let's do a gentle start and use the random library generate 5000 data points from a Gaussian distribution with $\mu = 2$ and $\sigma = 0.125$.**

```python
np.random.seed(42)
mu, sigma = 2, 0.125
s = np.random.normal(mu, sigma, size=5000)
```

**Now, let's use numpy.histogram to bin those number into 10 bins. What does the numpy.histogram function return? Do the two arrays have the same length?**

```python
counts = np.histogram(s,bins=10)[0]
bins = np.histogram(s,bins=10)[1]
```

The numpy.histogram function returns the values of the histogram and the values position.

```python
print(f"The counts have length {len(counts)} and the bins have length {len(bins)}.")
```

**Then we use matplotlib.pyplot.plot to plot the binned data. You will have to deal with the fact that the counts- and bin-arrays have different lengths. Explain how you deal with this problem and why.**

Only the first 10 bins are plotted. 

```python
x = np.repeat(bins, 2)[1:-1]
y = np.repeat(counts, 2)
plt.plot(x,y)
plt.xlabel('Bins')
plt.ylabel('Counts')
plt.show()
```

As expected, the histogram clearly shows that the data points are from a Gaussian distribution.

### Binning integers

**But binning real numbers into a fixed number of bins is easy when numpy.histogram does all the work and finds the right bin boundaries for you. Now we'll generate a bunch of integers and set the bin boundaries manually. This time, let's grab data from a Poisson distribution. As it turns out numpy also has some convenient random number generators. Use numpy.random.poisson to generate 5000 numbers drawn from a Poisson distribution characterized by $\lambda = 10$. Find the maximum and minimum value of your 5000 random numbers.**

```python
np.random.seed(42)
p = np.random.poisson(lam=10, size=5000)
```

```python
print(f"The maximum value is {max(p)} and the minimum value is {min(p)}.")
```

**Instead of simplify specifying the number of bins for numpy.histogram, let's specify the bins we want using a vector. Create a vector $v$ that results in a binning that puts each integer value in its own bin and where the first bin contains the minimum number you found above, and the last bin contains the maximum number (you may want to think about how numpy.histogram deals with the smallest an largest value in particular; is it correct to set the first entry of $v$ to be equal to the smallest value you observe and the last entry of $v$ to be the highest value observed?). Use the vector by setting numpy.histogram's bin parameter as bin = $v$. What is the sum over bin counts? Explain how the binning-vectors first and last element relates to the min and max from the Poisson distribution.**

```python
v = np.arange(min(p),max(p)+1)
#print(v)
counts, bins = np.histogram(p,bins=v)
counts, bins
```

```python
print(f"The sum of the bin counts is {sum(bins)}")
```

```python
plt.bar(v[:-1],counts)
plt.xlabel('Bins')
plt.ylabel('Counts')
plt.show()
```

### Binning and plotting degree distributions

**Let's generate the Erdös-Renyi (ER) network which has a degree distribution that matches the Poisson distribution above. First we have to figure out which values the ER parameters (N and p) should assume. It's easy to see that $N = 5000$, but how do you find $p$? Hint: The parameter $\lambda$ in the Poisson distribution corresponds to the average degree, so you have to find a $p$ that results in an average degree, $k = 10$. And you know that $\langle k \rangle = p (N-1)$, which will give you $p$.**

From $\langle k \rangle = p (N-1)$ $p$ can be isolated and results in the following expression:
$$p = \frac{\langle k \rangle}{N-1}$$
which for this specific case becomes
$$p = \frac{10}{4999}.$$

**Now, use networkx to create the graph and extract the degree distribution.**

```python
G = nx.erdos_renyi_graph(5000, (10/4999))
```

**Finally, create a nice bar plot of the degree distribution, including axes labels and a plot title. Make sure that it looks like the Poisson distribution you plotted above.**

```python
degree_dist = np.array([e for n, e in G.degree()])

d = np.arange(min(degree_dist),max(degree_dist)+2)

counts, bins = np.histogram(degree_dist,bins=d-(1/2))
plt.bar(d[:-1],counts)
plt.xlabel('Bins')
plt.ylabel('Counts')
plt.title('The degree distribution of the Erdös-Renyi network')
plt.show()
```

### In your notebook, work through NS exercise 3.1 ('Erdős-Rényi Networks'). The exercise can be found in Section 3.11: Homework.

Consider an Erdős-Rényi network with $N = 3000$ nodes, connected to each other with probability $p = 10^{–3}$.

**What is the expected number of links, $\langle L \rangle$?**

The expected number of links is determined using the following formula $$\langle L \rangle = \frac{pN(N-1)}{2}.$$

```python
N = 3000
p = 10e-3

E_L = (p*N*(N-1))/2
print(f"The expected number of links for the Erdös-Rényi network is {E_L}.")
```

**In which regime is the network?**

In order to determine which regime the network belongs to, the average degree $\langle k \rangle$ must be calculated using $$p(N-1).$$

```python
k = p*(N-1)
print(f"The average degree of the network is {k}.")
```

The network belongs to the supercritical regime since $\langle k \rangle > 1$. It also belong to the connected regime as $\langle k \rangle > \ln(N).$

```python
np.log(N)
```

**Calculate the probability $p_c$ so that the network is at the critical point.**

The probability $p_c$ is given by $$p_c = \frac{1}{N-1}.$$

For the network to be at the critical point the average degree, $\langle k \rangle$, needs to be equal to 1. 

```python
print(f'For the network to be at the critical point, the probability must be {1/(N-1)}.')
```

**Given the linking probability $p = 10^{-3}$, calculate the number of nodes $N^{cr}$ so that the network has only one component.**

That the network only has one component means that it have entered the connected regime, where the giant component has absorbed all nodes and components and $p> \ln(N)/N$. 


```python
#idx = np.where(0.001 > val)[0][0:2]

N = np.arange(1,10000)
p_cr = 10e-3
value = np.log(N)/N

print(np.where(p_cr > value)[0][0:2])
```

The smallest value of $N^{cr}$ where the inequality is still true, is chosen - $N^{cr} = 647$.

**For the network in (d), calculate the average degree $\langle k^{cr} \rangle$ and the average distance between two randomly chosen nodes $\langle d \rangle$.**

The average degree is given by $$\langle k^{cr} \rangle = p^{cr}(N^{cr}-1)$$ and the average distance is defined as $$\langle d \rangle = \frac{\ln(N^{cr})}{\ln(\langle k^{cr} \rangle)}.$$

```python
N_cr = 647
k_cr = p_cr*(N_cr-1)

d_cr = np.log(N_cr)/np.log(k_cr)
print(f"The average degree of the network above is {k_cr} \nand the average distance between two randomly chosen nodes is {d_cr}.")
```

**Calculate the degree distribution $p_k$ of this network (approximate with a Poisson degree distribution).**

The degree distribution on Poisson form is defined as $$p_k = \exp(-\langle k\rangle)\frac{\langle k^{cr} \rangle^{k^{cr}}}{k^{cr}!}.$$ 

```python
import math
p_k = math.exp(-k_cr)*(k_cr**k_cr)/(np.math.factorial(round(k_cr)))
print(f'The degree distribution of the network above is {p_k}.')
```

### Paths. Plot a random network with 200 nodes and an average degree of 1.5. (I suggest using networkx.draw and reading the documentation carefully to get an overview of all the options and what they look like. For example, you may want to shrink the node size).

It is given that $N=200$ and $\langle k \rangle = 1.5$. To make a random network the probability, $p$, needs to be determnined. This is done from the expression for the average degree: $$\langle k \rangle = p(N-1)$$ $$1.5 = p(200-1) \Leftrightarrow p = 1.5/(200-1).$$

```python
N = 200
k = 1.5
p = k/(N-1)

G_er = nx.erdos_renyi_graph(N,p)
nx.draw_spring(
    G = G_er, 
    node_size = 75, 
    node_color = '#A0CBE2', 
    node_shape = 'o',
    edge_color = 'k',
    with_labels = False, 
    width = 1.5, 
    edgecolors='k') 
plt.draw()
```

From the visualization of the graph it is clearly seen that the component in the middle is the Giant Connected Component. This component is extracted in the following exercise. 

**Extract the Giant Connected Component, GCC. (Hint. You can use networkx.connected_component_subgraphs)**

```python
con_comp = sorted(nx.connected_components(G_er), key = len, reverse = True)

GCC = G_er.subgraph(con_comp[0]).copy()

# Drawing the GCC
nx.draw_spring(
    G = GCC, 
    node_size = 75, 
    node_color = '#A0CBE2', 
    node_shape = 'o',
    edge_color = 'k',
    with_labels = False, 
    width = 1.3, 
    edgecolors='k') 
plt.draw()
```

**Choose a node at random from the GCC. (Hint: You may want to try random.choice.)**

```python
random_node = np.random.choice(G_er.subgraph(GCC))
random_node
```

### **Find all nodes that are precisely 2 steps away from that node. (Hint. I suggest networkx.single_source_shortest_path_length)**

```python
max_two_steps = nx.single_source_shortest_path_length(GCC, random_node, cutoff = 2)
two_steps = np.array([node for (node, steps) in max_two_steps.items() if steps == 2])
print(f"The nodes that are precisely 2 steps from the random node are {two_steps}.")
```

**Plot the GCC with the following choice of colors. Starting node black ("#000000"). The nodes 2 steps away red ("#ff0000"). All other nodes blue ("#A0CBE2"). Again, I suggest using networkx.draw() and reading the documentation carefully find out how to color individual nodes.**

```python
nodes = np.array(list(GCC.nodes))
idx_random_node = np.where(nodes == random_node)[0]
idx_two_step = np.where(np.in1d(nodes,two_steps))[0]
```

```python
node_colors = np.array(['#A0CBE2'] * len(GCC))
node_colors[idx_random_node] = '#000000'
node_colors[idx_two_step] = "#ff0000"

nx.draw_spring(
    G = GCC, 
    node_size = 75, 
    node_color = node_colors, 
    node_shape = 'o',
    edge_color = 'k',
    with_labels = False, 
    width = 1.5, 
    edgecolors='k') 
plt.draw()
```
