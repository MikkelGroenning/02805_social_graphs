* Exercise 4
*Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men, women, and people in each document. What has happened to the usage of these words over time?*

````python
from nltk.corpus import state_union
````

````python
words = {
    'men' : 0,
    'women' : 0,
    'people' : 0,
}

for word in state_union.words():
    if word in words:
        words[word] += 1
````

````python
words = ['men', 'women', 'people']
cfd = nltk.ConditionalFreqDist(
    (target, file_id)
    for file_id in state_union.fileids()
    for w in state_union.words(file_id)
    for target in words
    if w.lower().startswith(target)
)
````

````python
plt.figure(figsize=(12, 8))
cfd.plot()
plt.show()
````

*Define a conditional frequency distribution over the Names corpus that allows you to see which initial letters are more frequent for males vs. females (cf. 4.4).*

````python
from nltk.corpus import names
cfd = nltk.ConditionalFreqDist(
    (fileid, name[0])
    for fileid in names.fileids()
    for name in names.words(fileid)
)
````

* Exercise 18

*Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.*

````python
bigram_generator = nltk.bigrams(brown.words())

bigrams = []
while len(bigrams) != 50:
    bigram = next(bigram_generator)

    potetial_bigram = set(word for word in bigram if word.lower() not in stopwords)
    if len(potetial_bigram) == 2:
        bigrams.append(potetial_bigram)
    else:
        continue
````

````python
import string
````

````python
text = brown.words()
bigrams = [w for w in nltk.bigrams([w.lower() for w in text]) 
               if not any(word in stopwords+list(string.punctuation) for word in w)]
bigrams = nltk.FreqDist(bigrams)
````

````python
bigrams.most_common(50)
````
