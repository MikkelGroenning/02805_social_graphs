>2. As you build the graph, store the attributes of the nodes (remember the `universe` from the past week). On top of that, let's also store the `length of the content` as one of the attributes of the node (i.e. count the number of *words* you have in each character's page). By storing this information as nodal features, we keep all the relevant data in one place. The safe way to count words without counting symbols is to use `regex`, but simpler methods are OK.

```python
import pandas as pd
import numpy as np
import json
import networkx as nx
import re
import matplotlib.pyplot as plt
import pickle
```

```python
with open('super_hero.json', 'r') as fp:
    super_hero_dict = json.load(fp)
```

```python
wiki_super_heroes = np.array(list(super_hero_dict.keys()))
```

```python
n = len(super_hero_dict)
adj_mat = np.zeros((n,n))
wiki_length = np.zeros(n)

for i, super_hero in enumerate(super_hero_dict.keys()):

    # check if super hero has wiki page
    if super_hero_dict[super_hero]:
         
        wiki_name = super_hero_dict[super_hero]['Wiki_Name']
        wiki_text = super_hero_dict[super_hero]['Wiki_text']
        
        # Get links in the superhero wiki:
        if wiki_text:
            links = re.findall("\[\[(.*?)[\||\]\]]", wiki_text)
            intersection = np.array(list(set(links) & set(wiki_super_heroes)))

            adj_mat[i, np.in1d(wiki_super_heroes, intersection)] = 1    

            # caluclate the length of the wiki text:
            wiki_length[i] = len(re.findall(r'\b[\w-]+\b', re.sub(r'(<ref.+?(<\/ref>|\/>)|\n)', "", wiki_text)))

#    if i == 2967:
#        break
```

```python
super_hero_graph_directed = nx.from_numpy_matrix(adj_mat, create_using=nx.DiGraph)

label_mapping = {idx:key for idx, key in enumerate(super_hero_dict.keys())}
super_hero_graph_directed = nx.relabel_nodes(super_hero_graph_directed, label_mapping)
```

```python
for i, node in enumerate(super_hero_graph_directed.nodes):
    super_hero_graph_directed.nodes[node].update({
        "Universe" : super_hero_dict[node]['Universe'],
        "length_of_content" : wiki_length[i]
        })
```

```python
nx.write_gpickle(super_hero_graph_directed, 'super_hero_graph_directed.pickle')
```

*Exercise 2:* Let's build a simple visualization of network
>1. For the sake of the visualisation, let's convert our network to *undirected graph* (tip: There is a NetworkX command to help you).
>   * *Note*: Keep the directed graph, we will use it in couple of the following exercises
>2. Use the NetworkX command `nx.draw_kamada_kawai` to draw the network resulting undirected. (You can find [background on the algorithm here](https://en.wikipedia.org/wiki/Force-directed_graph_drawing).)  
>   * Node color should depend on the `universe` attribute (i.e. use different colors for DC and Marvel universe)
>   * Set up your plot so that node-size depends on the node *degree*
>   * **Optional**: Set up you plot so that edge color depends on the `universe` attribute. If an edge connects two nodes from Marvel universe assign `color_a`, if the edge connects two nodes from DC universe assign `color_b`, in case a link connects nodes from disparate universes, cases assign `color_c`. (Here, you get to decide on the colors.)

```python
super_hero_graph = super_hero_graph_directed.to_undirected()

con_comp = sorted(nx.connected_components(super_hero_graph), key = len, reverse = True)
GCC = super_hero_graph.subgraph(con_comp[0]).copy()
```

```python
node_colors = ['tab:red' if universe == 'Marvel' else 'tab:blue'  
    for super_hero, universe in nx.get_node_attributes(GCC, 'Universe').items()]
```

```python
node_sizes = [degree*2 for degree in dict(GCC.degree()).values()]
```

```python
def color_edge(unv1, unv2):
    if unv1 == 'Marvel' and unv2 == 'Marvel':
        color = 'tab:orange'
    elif unv1 == 'DC' and unv2 == 'DC':
        color = 'tab:cyan'
    else:
        color = 'tab:purple'
    return color 

edge_colors = [
    color_edge(*(nx.get_node_attributes(GCC, 'Universe')[f],  nx.get_node_attributes(GCC, 'Universe')[t]))
    for f, t in GCC.edges
]
```

```python
plt.figure(figsize=(14,10))
nx.draw_kamada_kawai(
    G = GCC, 
    node_size = node_sizes,
    node_color = node_colors, 
    node_shape = 'o',
    edge_color = edge_colors,
    with_labels = False, 
    width = 0.5,
    edgecolors = 'k',
    alpha = 0.5,
)
plt.show()
```

```python
from fa2 import ForceAtlas2
```

```python
forceatlas2 = ForceAtlas2(
    # Behavior alternatives
    outboundAttractionDistribution=True,  # Dissuade hubs
    linLogMode=False,  # NOT IMPLEMENTED
    adjustSizes=False,  # Prevent overlap (NOT IMPLEMENTED)
    edgeWeightInfluence=1.0,

    # Performance
    jitterTolerance=1.0,  # Tolerance
    barnesHutOptimize=True,
    barnesHutTheta=1.2,
    multiThreaded=False,  # NOT IMPLEMENTED

    # Tuning
    scalingRatio=2.0,
    strongGravityMode=False,
    gravity=1.0,

    # Log
    verbose=True
)

plt.figure(figsize=(14,10))
positions = forceatlas2.forceatlas2_networkx_layout(GCC, pos=None, iterations=2000)
nx.draw_networkx_nodes(
    G = GCC,
    pos = positions, 
    node_size=20,
    node_color=node_colors,
    alpha=0.4
)
nx.draw_networkx_edges(
    G = GCC,
    pos = positions, 
    edge_color = edge_colors, 
    alpha=0.05
)
plt.axis('off')
plt.show()
```
